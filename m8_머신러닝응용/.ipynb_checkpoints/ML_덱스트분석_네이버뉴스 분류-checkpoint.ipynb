```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
from selenium.webdriver.common.keys import Keys
import time
path = 'G:/내 드라이브/tool/chromedriver.exe'
driver = webdriver.Chrome(path)
driver.get("https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=100")
list=[]
page=1
for _ in range(10):
    for step in range(2,11):
        for i in range(1,5):
            for j in range(1,6):
                try:
                    article=f'//*[@id="section_body"]/ul[{i}]/li[{j}]/dl/dt[2]/a'
                    print(article)
                    driver.find_element_by_xpath(article).send_keys(Keys.ENTER)
                    text=driver.find_element_by_xpath('//*[@id="articleBodyContents"]')
                    list.append(text.text)
                    driver.back()
                except: 
                    continue
        print(f'{page}페이지 완료') 
        driver.find_element_by_xpath(f'//*[@id="paging"]/a[{step}]').send_keys(Keys.ENTER)
        page += 1
        address=f'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=100#&date=%2000:00:00&page={page}'
        print(address)
        driver.get(address)
        time.sleep(1)
    driver.find_element_by_xpath('//*[@id="paging"]/a[10]').send_keys(Keys.ENTER)
    driver.get(f"https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=100#&date=%2000:00:00&page={page}")
    time.sleep(1)
```
