```python
import seaborn as sns
import numpy as np
dft=sns.load_dataset('titanic')
dft
```

```python
dft2=dft.drop(dft.loc[dft.survived==0].index)
dft['family']=dft.sibsp+dft.parch
dft.drop(['sibsp','parch'],axis=1)
dft3=dft2.drop(['alone','alive','fare','embarked','embark_town','class','who','adult_male'],axis=1)
dft3.index=np.arange(len(dft3))
print(dft3.age[2])
for i in range(len(dft3)):
    if dft3.age[i] < 20:
        dft3.age[i]=20
    elif dft3.age[i] < 30:
        dft3.age[i] = 30
    elif dft3.age[i] < 40:
        dft3.age[i] = 40
    elif dft3.age[i] < 50:
        dft3.age[i] = 50
    else:
        dft3.age[i] = 60
dft3
```
