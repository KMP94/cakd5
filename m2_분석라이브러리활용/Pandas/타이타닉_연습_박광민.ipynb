```python
import seaborn as sns
import numpy as np
dft=sns.load_dataset('titanic')
dft
```

```python
#[과제] 데이터 추출(인덱싱, 조건),, 변경 ,삭제 ,정렬, 칼럼 병합(sibsp+parch=family) 등을 수행하여 데이터에서 인사이트를 찾아보세요
#과대 적합(과대한 중복 정보)
# dft2=dft.drop(dft.loc[dft.survived==0].index)
# dft['family']=dft.sibsp+dft.parch
# dft2=dft.drop(['alone','alive','class','who','adult_male','sibsp','parch'],axis=1)
dft3=dft.drop(dft.loc[dft.survived==0].index)
# dft['family']=dft.sibsp+dft.parch

dft3.index=np.arange(len(dft3))
print(dft3.age[2])
for i in range(len(dft3)):
    if dft3.age[i] < 20:
        dft3.age[i]=20
    elif dft3.age[i] < 30:
        dft3.age[i] = 30
    elif dft3.age[i] < 40:
        dft3.age[i] = 40
    elif dft3.age[i] < 50:
        dft3.age[i] = 50
    else:
        dft3.age[i] = 60
dft3
```
