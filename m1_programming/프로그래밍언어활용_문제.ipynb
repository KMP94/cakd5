#### 11. 정규표현식을 사용하여 text에서 지역코드만 출력하세요.
text = "문의사항이 있으면 032-232-3245 으로 연락주시기 바랍니다."

#### 12. 'abcdefghij' 에 대하여 중첩을 적용한 서브그룹 5개로 컴파일하여 group()함수를 이용하여 'abcdefghi'와 e'를 출력하세요

```python
import re
text = "abcdefghij"
pattern=re.compile('((a(bc)d(e)f(g)hi)j)')
m=pattern.search(text)
print(m.group(2))
print(m.group(4))
```

####  16. numbers 내부에 들어 있는 숫자가 몇 번 등장하는지를 출력하는 코드를 작성하세요.
numbers = [3,6,3,5,8,2,9,7,4,3,6,3,5,8,2,9,7,4]

```python
numbers = [3,6,3,5,8,2,9,7,4,3,6,3,5,8,2,9,7,4]
dic={}
for num in set(numbers):
    if num in numbers:
        dic[num] = numbers.count(num)
print(dic)
```

#### 18. 사칙연산을 수행하는  클래스를 작성하여 객체를 만들고 2개의 정수 10, 5를 입력한 후 연산 결과를  포멧을 이용하여 출력하세요.

```python
class Cal:
    def __init__(self,num1,num2):
        self.num1=num1
        self.num2=num2
    def add(self):
        return self.num1+self.num2
    def sub(self):
        return self.num1-self.num2
    def mul(self):
        return self.num1*self.num2
    def div(self):
        try:
            return self.num1/self.num2
        except:
            print('0으로 나눌수 없습니다.')
try:
    num1=int(input())
    num2=int(input())
except:
    print('정수를 입력하세요')
else:
    cal=Cal(num1,num2)
    print(f'add: {cal.add()}, sub:{cal.sub()}, mul:{cal.mul()}, div: {cal.div()}')
```
