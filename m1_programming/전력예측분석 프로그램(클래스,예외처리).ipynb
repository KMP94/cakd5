```python
class CheckError:
    def check_num_range(self,start,end):
        while 1:
            try:           
                num=int(input())
                if start<= num <= end:
                    return num
                    break
                else:
                    print(f'{start}에서 {end} 사이의 값을 입력하세요. :',end=' ')
            except:
                print(f'{start}에서 {end} 사이의 값을 입력하세요. :',end=' ')
    def check_pnum(self):
        while 1:
            try:
                num=int(input())
                if num > 0:
                    return num
                    break
                else:
                    print(f'양수를 입력하세요. :',end=' ')
            except:
                print(f'0이상의 정수를 입력하세요. :',end=' ')
    def check_chr(self):
        while 1:
            try:
                chr=input()
                if chr=='Y' or chr=='N' or chr=='y' or chr=='n':
                    return chr
                    break
                else:
                    print('Y/N 중에서만 선택하세요.',end=' ')
            except:
                print('Y/N 중에서만 선택하세요.',end=' ')
class Electric:
    def billed(self,ele,tex):
        result=ele+tex
        tmp=result%10
        return result-tmp
    def tex_price(self,ele_price):
        result=int(ele_price*0.1)  #부가세
        result1=int(ele_price*0.037)  #기반기금
        temp=result1
        temp=temp%10
        result1=result1-temp
        return result+result1
    def ele_price(self,amount):
        if amount==0:
            return print("사용량이 없습니다.",end=' ')
        elif amount <=200:
            b=910
            y=88.3
            self.ele_p=b+int(amount*y)+int(amount*5.3)
        elif 200<amount<=400:
            b=1600
            y=182.9
            self.ele_p= b+int((amount-200)*y)+int(200*88.3)+int(amount*5.3)
        elif 400<amount<=1000:
            b=7300
            y=275.6
            self.ele_p= b+int((amount-400)*y)+int(200*182.9)+int(200*88.3)+int(amount*5.3)
        else:
            y=704.5
            self.ele_p= int((amount-1000)*704.5)+int(600*275.6)+int(200*182.9)+int(200*88.3)+int(amount*5.3)
        return self.ele_p
class Machine:
    def set_name(self):
        self.m_name=input("제품 이름 : ")
    def start_time(self):
        print(f'제품의 시작시간을 입력하세요. :',end=' ')
        self.s_time=CheckError.check_num_range(self,0,24)
    def end_time(self):
        print(f'제품의 종료시간을 입력하세요. :',end=' ')
        self.e_time=CheckError.check_num_range(self,0,24)
    def set_cost(self):
        print('시간당 소비 전력량을 입력해 주세요(단위:kwh)',end=' ')
        self.cost=CheckError.check_pnum(self)
    def using_time(self):
        if (self.e_time-self.s_time)>0:
            self.u_time = self.e_time-self.s_time
        elif (self.e_time-self.s_time)<0:
            self.u_time=self.e_time-self.s_time+24
        else:
            self.u_time=0
    def get_amount(self):
        self.amount=self.u_time*self.cost
        return self.amount
    def machine_print(self):
        print(f'제품이름 : {self.m_name}, 사용시간 : {self.u_time}, 사용전기량 : {self.amount}')
    def get_am(self,num1,num2):
        am=0
        if num1<=12 and num2<=12:
            if num2>num1:
                am=num2-num1
            elif num1>num2:
                am=(12-num1)+num2
            else:
                am=0
        elif num1 <=12 and num2>=13:
            am=12-num1
        elif num1>=13 and num2>=13:
            if num1<num2:
                am=0
            elif num2<num1:
                am=12
            else:
                am=0
        elif num1>=12 and num2<=12:
            am=num2
        else:
            am=0
        return am
    def get_pm(self,num1,num2):
        if num1>=12 and num2>=12:
            if num2>num1:
                pm=num2-num1
            elif num1>num2:
                pm=(24-num1)+(num2-12)
            else:
                pm=0
        elif num1 <= 12 and num2 >=13:
            pm=num2-12
        elif num1>13 and num2<=12:
            pm=24-num1
        else:
            pm=0
        return pm
class SetFrame:
    
#     def get_m_list(self):
#         return self.m_list
    def main_f(self):
        print('''
*********************************************************************

                    현재 페이지는 설정 페이지 입니다.                         

          (1):설정          (2):초기화         (3):뒤로가기

*********************************************************************
        ''')

        while 1:
            print('메뉴선택 :',end='')
            choice=CheckError.check_num_range(self,1,4)
            if choice==1:
                while 1:
                    self.op=True
                    m=Machine()
                    m.set_name()
                    m.start_time()
                    m.end_time()
                    m.set_cost()
                    m.using_time()
                    m.get_amount()
                    self.m_list.append([m.m_name,m.s_time,m.e_time,m.cost,m.amount])
                                        # 0-이름 1-시작시간 2-종료시간 3-비용 4-사용량
                    print('추가 등록 하시겠습니까?[Y/N] :',end=' ')
                    ch=CheckError.check_chr(self)
                    if ch=='Y' or ch=='y':
                        print('다시 입력 합니다. ') 
                    else:
                        print('(1):설정   (2):초기화   (3):뒤로가기')
                        break
            elif choice==2:
                self.m_list=[]
                print('데이터가 초기화 되었습니다.')
            elif choice==3:
                MainFrame.main_f(self)
class analyzeFrame:
    def main_f(self):
        print('''
*********************************************************************

                    설정 값에 대한 분석 페이지 입니다

   (1):일반 분석   (2):오전/오후 통계    (3):시나리오   (4):뒤로가기

*********************************************************************     
      ''')
        while 1:
            print('메뉴선택 :',end='')
            # self.m_list.append([m.m_name,m.s_time,m.e_time,m.cost,m.amount])
            choice=CheckError.check_num_range(self,1,5)
            if choice==1:
                for i in range(0,len(self.m_list)):
                    e_price=Electric.ele_price(self,self.m_list[i][4])
                    t_price=Electric.tex_price(self,e_price)
                    print(f'이름:{self.m_list[i][0]}, 전기요금:{e_price}, 세금:{t_price}, 청구액:{Electric.billed(self,e_price,t_price)}')    
                print()
                print(f'(1):일반 분석, (2):오전/오후 통계, (3):시나리오, (4):뒤로가기')
            elif choice==2:
                am_full_run=0
                pm_full_run=0
                am_full_amount=0
                pm_full_amount=0
                for i in range(0,len(self.m_list)):
                    name=self.m_list[i][0]
                    run_am=Machine.get_am(self,self.m_list[i][1],self.m_list[i][2])
                    run_pm=Machine.get_pm(self,self.m_list[i][1],self.m_list[i][2])
                    cost=self.m_list[i][3]
                    am_full_run += run_am
                    pm_full_run += run_pm
                    am_full_amount += (cost*run_am)
                    pm_full_amount += (cost*run_pm)              
                    print(f'이름 : {name}, 오전 사용 시간 : {run_am}, 오후 사용 시간 : {run_pm}')
                print()
                print(f'(1):일반 분석, (2):오전/오후 통계, (3):시나리오, (4):뒤로가기')
                am_e_price=Electric.ele_price(self,am_full_amount)
                am_t_price=Electric.tex_price(self,am_e_price)
                pm_e_price=Electric.ele_price(self,pm_full_amount)
                pm_t_price=Electric.tex_price(self,pm_e_price)
                print(f'(오전) 시간 : {am_full_run}, 요금 : {Electric.billed(self,am_e_price,am_t_price)}, 전력소모량 :{am_full_amount}')
                print(f'(오후) 시간 : {pm_full_run}, 요금 : {Electric.billed(self,pm_e_price,pm_t_price)}, 전력소모량 :{pm_full_amount}')
                
                print(f'(1):일반 분석, (2):오전/오후 통계, (3):시나리오, (4):뒤로가기')
            elif choice==3:
                print('현재 상태에서 한달안에 사용될 가상의 사용일을 입력하세요( 최대 : 365 ) ) : ')
                day=CheckError.check_num_range(self,1,365)
                tmp=0
                for i in range(0,len(self.m_list)):
                    tmp+=self.m_list[i][4]
                total_amount=tmp*day
                e_price=Electric.ele_price(self,total_amount)
                t_price=Electric.tex_price(self,e_price)
                print(f'30일동안 현재 사용량 기준{day}일 사용시 소모전력량은 "{total_amount}"kwh 이며, 청구액은 "{Electric.billed(self,e_price,t_price)}"원 입니다.')
                print(f'(1):일반 분석, (2):오전/오후 통계, (3):시나리오, (4):뒤로가기')
            elif choice==4:
                MainFrame.main_f(self)

class MainFrame:
    m_list=[]
    op=False
    def main_f(self):
        print('''
*********************************************************************

                     전력 분석 예측 프로그램 입니다.
                  (초기 실행시 설정먼저 부탁드립니다. )
          (1):설정           (2):분석           (3):종료

*********************************************************************     
        ''')
        while 1:
            print('메뉴선택 :',end='')
            choice=CheckError.check_num_range(self,1,3)
            if choice==1:
                SetFrame.main_f(self)
            elif choice==2:
                if self.op==True:
                    analyzeFrame.main_f(self)
                else:
                    print('분석을 실행할 데이터가 존재하지 않습니다.')
            elif choice==3:
                break
        print('분석 예측 프로그램을 종료 합니다.')
    

start=MainFrame()
start.main_f()
```
