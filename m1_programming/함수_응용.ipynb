```python
# 수학 ! factorial 사용법
def factorial(n):
    result=1
    for i in range(1,n+1):
        result*=i
    return result
factorial(10)
```

```python
# 재귀란 자기 자신을 호출 하는것
#재귀함수를 이용해서 팩토리얼을 구하는 사용자 함수를 작성하고 1!,5!, 10!를 구하세요
#factorial(n)=n*factorial(n-1)(n>0)
def factorial(n):
    result=1
    if n-1>=0:
        result=n*factorial(n-1)
    return result
factorial(0)
```

```python
#정규표현식 분석 필요 이해
import re
states = ['Alabama', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda','south carolina', 'West virginia']
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip() # 양쪽 공백제거
        value = re.sub('[!#?]','',value) # 특수문자 제거  대체 앞에 파라미터를 공백으로 대체
        value = value.title() # 첫 글자만 대문자로 변경
        result.append(value)
    return result

clean_strings(states)
```

```python
from os.path import exists

def getTop10BabyName(year):
    nameF = {}
    nameM = {}
    
    filename = 'data/names/yob%s.txt'%year
    if not exists(filename):
        print('[%s] 파일이 존재하지 않습니다.'%filename)
        return None
    
    with open(filename,'r') as f:
        data = f.readlines()
        for d in data:
            if d[-1] == '\n':
                d = d[:-1]
                
                tmp = d.split(',')
                name = tmp[0]
                sex = tmp[1]
                birth = tmp[2]
                
                if sex == 'F':
                    ret = nameF
                else:
                    ret = nameM
                    
                if name in ret:
                    ret[name] += int(birth)
                else:
                    ret[name] = int(birth)
                    
    retF = sorted(nameF.items(), key=lambda x:x[1],reverse=True)
    retM = sorted(nameM.items(), key=lambda x:x[1], reverse=True)

    for i, name in enumerate(retF):
        if i>9:
            break
        print('TOP_%d 여자아기이름:%s'%(i+1,name))
    for i, name in enumerate(retM):
        if i>9:
            break
        print('Top_%d 남자아기이름:%s'%(i+1,name))
y = input('출생년도 입력하세요(예:2001):')
getTop10BabyName(y)
```

```python
import time
def long_task(): #5초의 시간이 걸리는 함수
    for i in range(5):
        time.sleep(1)
        print(f'working : {i}\n')
print('Start')
for i in range(5):
    long_task()
print("End")
```

```python
import time
import threading
def long_task():
    for i in range(5):
        time.sleep(1)
        print(f'working : {i}\n')
print('Start')
threads = []
for i in range(5):
    t=threading.Thread(target=long_task)
    threads.append(t)
for t in threads:
    t.start()
for t in threads:
    t.join() #join으로 스레드가 종료될 때까지 기다린다.
print('End')
```
