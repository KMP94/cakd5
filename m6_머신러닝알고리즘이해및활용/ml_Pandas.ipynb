```python
# [과제] t_df의 name 컬럼을 이용하여 각 name의 알파벳 글자수를 표현하는 새로운 컬럼 Name_len을 출력하세요(람다식 이용)
import re
t_df1=t_df.copy()

t_df1['Name_len'] = t_df1['Name'].apply(lambda x: len(re.sub('[()".,\ ]','',x)))
t_df1[['Name','Name_len']]
```

```python
# [과제] t_df의 Age 컬럼을 이용하여 새로운 컬럼 Age_cat를 사용자 함수를 이용하여 출력하세요(7개 카테고리)
def get_category(age):
    cat = ''
    if age < 10 : cat = 'Child'
    elif 10 <= age < 20: cat = 'Teenage'
    elif 20 <= age < 30: cat = 'Twenty'
    elif 30 <= age < 40: cat = 'Thirty'
    elif 40 <= age < 50: cat = 'Forty'
    elif 10 <= age < 60: cat = 'Fifty'
    else : 
        cat = 'Elderiy'
    return cat
t_df['Age_cat'] = t_df.Age.apply(lambda x:get_category(x))
t_df[['Age','Age_cat']].head()
```

```python
# [과제] t_df에 대하여 결손값 처리, 불필요한 컬럼 삭제, 인코딩 등 을 수행하는 전처리 과정을 
# 하나의 사용자 함수(def transform_features(df))로 작성하여 분석용 데이터셋을 만드세요.
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
def transform_features(df):
    #파생변수 생성.
    #결손값
    df['Age'].fillna(df['Age'].mean(), inplace=True)
    df['Cabin'].fillna('N', inplace=True)
    df['Embarked'].fillna('N', inplace=True)
    df['Fare'].fillna(0, inplace=True)
    #삭제
    df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)    
    #인코딩
    df['Cabin'] = df['Cabin'].str[:1]
    features = ['Cabin', 'Sex', 'Embarked']
    for feature in features:
        le = LabelEncoder()
        le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    return df
t_df1=t_df.copy()
transform_features(t_df1)
t_df1
```

```python
# [과제] t_df Name 컬럼에서 Titel(칭호)를 추출해서 이에 기반한 age 평균값을 이용하여 age null 처리를 한 후 개선된 부분을 설명하세요.
t_df1=t_df
t_df1['Title']=''
for i in range(len(t_df1['Name'])):
    t_df1['Title'][i]=t_df1['Name'][i].split(',')[1].split('.')[0].lower()
t_df1

t_df1['Age']=t_df1.groupby('Title').Age.apply(lambda x:x.fillna(x.mean()))
t_df1.info()

# 일괄적인 평균값보다 특정지위별 평균으로 접근하기에 좀더 정확도가 높아지리라 기대함.
```
